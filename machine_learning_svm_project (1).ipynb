```python
x=a.drop("quality",axis=1)  #feature col
y=a["quality"]  # target col
```

```python
a.groupby('quality').mean()
```

```python
ss=StandardScaler()  # scaling
```

```python
x=ss.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2329)      # splitting data into training and testing
```

```python
model=SVC()      # deploy model
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)     # predict model
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
print("classification_report",classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))          # evaluate model
```
